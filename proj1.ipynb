```
import pandas as pd
import matplotlib.pyplot as plt
```

```
from google.colab import files
import os

# Check if the file already exists to avoid re-uploading
if not os.path.exists('netflix_titles.csv'):
  print('Please upload the netflix_titles.csv file:')
  uploaded = files.upload()
  if 'netflix_titles.csv' in uploaded:
    print('netflix_titles.csv uploaded successfully.')
  else:
    print('Failed to upload netflix_titles.csv. Please try again.')
else:
  print('netflix_titles.csv already exists.')
```

```
df = pd.read_csv("netflix_titles.csv")

print(df.head())
```

```
print(df.info())
print(df.shape)
```

```
print(df.isnull().sum())
```

```
type_count = df["type"].value_counts()

type_count.plot(kind="bar")

plt.title("Movies vs TV Shows")
plt.show()
```

```
country = df["country"].dropna()

country = country.str.split(",").explode()

country.value_counts().head(10).plot(
    kind="bar"
)

plt.title("Top Countries")
plt.show()
```

```
df["date_added"] = pd.to_datetime(
    df["date_added"],
    errors="coerce"
)

df["year"] = df["date_added"].dt.year

df["year"].value_counts()\
.sort_index().plot()

plt.title("Content Added by Year")

plt.show()
```

```
df["rating"].value_counts()\
.head(10).plot(
    kind="bar"
)

plt.title("Ratings")

plt.show()
```
